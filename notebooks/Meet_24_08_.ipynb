```python
# Import libraries

import matplotlib.pyplot as plt

from braindecode.datasets import MOABBDataset
from braindecode.preprocessing import \
    create_windows_from_events, create_fixed_length_windows
from braindecode.preprocessing import preprocess, Preprocessor
import seaborn as sns

# Definindo cores do plot
my_colors = plt.rcParams['axes.prop_cycle']()
```

```python
# Load dataset
dataset = MOABBDataset(dataset_name="BNCI2014001", subject_ids=[1])
```

```python
# Checking the type of dataset
type(dataset.description)
```

```python
preprocessors = [
    Preprocessor('pick_types', eeg=True, meg=False, stim=True),
    Preprocessor('resample', sfreq=100)
]
print(dataset.datasets[0].raw.info["sfreq"])
preprocess(dataset, preprocessors)
print(dataset.datasets[0].raw.info["sfreq"])
```

```python
# Pegando o dataset
df_raw = dataset.datasets[0].raw.to_data_frame()
```

```python
# Criando funcão de Plot

def plot_eeg(df, labels = [""], time_step = 0.01):

  # Configura timestep de ms para step do dataframe
  df_step = int(time_step*100)

  # Filtra o dataframe por step
  df = df.iloc[::df_step].reset_index(drop = True)

  # Contagem do numero de eletrodos escolhidos
  num_labels = len(labels)

  # Se apenas 1 eletrodo for escolhido, fazer plot
  if num_labels == 1:

    plt.figure(figsize = (20,3))
    sns.lineplot(x = df["time"], y = df[labels[0]])

    plt.title("Eletrodo "+ labels[0])

  # Se mais de 1 eletrodo
  if num_labels > 1:

    # Ajuste do numero de linhas e tamanho
    fig, axes = plt.subplots(nrows=num_labels, ncols=1, figsize=(20,3))
    plt.subplots_adjust(hspace=0)

    for i in range(len(labels)):
      ax = sns.lineplot(x = df["time"], y = df[labels[i]], ax=axes[i],**next(my_colors))

plot_eeg(df    = df_raw,
         labels = ["C1", "C4"],
         time_step = 0.01)
```

```python
colunas = ['C5', 'C3', 'C1', 'Cz', 'C2', 'C4', 'C6']
```

```python
plot_eeg(df    = df_raw,
         labels = colunas,
         time_step = 0.5)
```
